##VISUAL


```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[5,6,7,8]
plt.plot(x,y,color="red",marker="s",linestyle="--",markersize="4",linewidth="1")#marker star
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13,7))#height,width

x1=[1,2,3,4]
x2=[5,6,7,8]
x3=[9,10,11,12]
x4=[3,6,8,9]

plt.subplot(2,2,1)
plt.plot(x1,x2)
plt.title("first plot")

plt.subplot(2,2,2)
plt.plot(x2,x3)
plt.title("second plot")

plt.subplot(2,2,3)
plt.plot(x2,x1)
plt.title("third plot")

plt.subplot(2,2,4)
plt.plot(x3,x4)
plt.title("fourth plot")
plt.tight_layout()
plt.show()
```

```python
#barplot-supports category and values
category=["a","b","c","d"]
values=[10,20,30,40]

plt.bar(category,values,color="purple")
```

```python
#Histogram-used by distrirbution values with bins

x=[1,1,2,2,4,5,3,3,1,4,3]
plt.hist(x,bins=5)
plt.show()
```

```python
#Scatter plot-understand data interconnection and distribution

x=[1,2,3]
y=[4,5,6]

plt.scatter(x,y,color="orange")
plt.show()
```

```python
#piechart-each label has size
size=[10,20,30,40]
labels=["book","pen","led","eraser"]
colors=["red","blue","green","yellow"]
explode=(0.4,0,0,0)

plt.pie(size,labels=labels,colors=colors,explode=explode)
plt.show()
```

Seaborn has dataset which is tips


```python
import seaborn as sns
t=sns.load_dataset("tips")
t.head()
```

```python
sns.scatterplot(x="total_bill",y="tip",data=t)
plt.title("Scatter Plot")
```

```python
sns.lineplot(x="size",y="total_bill",data=t)
```

```python
sns.barplot(x="day",y="tip",data=t)
```

```python
sns.boxplot(x="day",y="tip",data=t)
```

```python
sns.histplot(t["total_bill"],bins=5,kde=True)
```

```python
sns.kdeplot(t["total_bill"],shade=True)
```

```python
sns.pairplot(t)
```

```python
corr=t[["total_bill","tip","size"]].corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
